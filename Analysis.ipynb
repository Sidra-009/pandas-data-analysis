```python
import pandas as pd
```

```python
data=pd.read_csv('DataScience_salaries_2024.csv')
```

```python
data
```

```python
#in sql we call it join
```

```python
sample = data.sample(1000)   # take only 100 rows
merged = sample.merge(sample, left_on='company_location', right_on='employee_residence')
```

```python
merged
```

```python
merged.shape
```

```python
print(sample.shape)
print(merged.shape)
```

```python
merged.groupby('job_title_x')['remote_ratio_x'].sum()
```

```python
#multi index seeries
merged.groupby(['work_year_x','job_title_x'])['salary_x'].sum()
```

```python
type(merged.groupby(['work_year_x','job_title_x'])['salary_x'].sum())
```

```python
# reset_index conevert seroes into dataframe
merged.groupby(['work_year_x','job_title_x'])['salary_x'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='job_title_x',keep='first').sort_values('work_year_x')
```
